# Algorithme Hastings-within-Gibbs

Le code suivant implémente un algorithme de chaîne de Markov pour ajuster un modèle de régression logistique mixte bayésienne. La régression logistique mixte est une technique statistique couramment utilisée pour analyser des données longitudinales ou des données de survie. Elle permet de modéliser la probabilité de succès ou d'échec d'un événement en fonction de variables explicatives continues ou catégorielles, tout en prenant en compte l'effet de variables latentes ou cachées. L'algorithme de chaîne de Markov est utilisé pour échantillonner de manière itérative les paramètres du modèle et estime la distribution a posteriori des paramètres. Ce code est une implémentation pratique de cette technique et permet de modéliser des données à l'aide de la régression logistique mixte bayésienne.

```python
def oxford(nchain, initial_alpha, initial_beta1, initial_beta2, initial_sigma, initial_mu, initial_b, proposition_sd, K, r1, n1, r0, n0, year):
    
    
    ## Cette fonction utilise l'échantillonneur Hastings-within-Gibbs pour simuler alpha, beta1, beta2 et sigma**2
    
    ## nchain: taille de la chaine
    ## initial_alpha, initial_beta1, initial_beta2, initial_sigma, initial_mu, initial_b: valeurs initiales
    ## proposition_sd: les variances pour les marches aléatoires des propositions
    ## K: la taille des données
    ## r1, n1, r0, n0, year: les données
    
    alpha = initial_alpha
    beta1 = initial_beta1
    beta2 = initial_beta2
    sigma = initial_sigma
    mu = initial_mu
    b = initial_b 
    
    log_psi = alpha + beta1 * year + beta2 * (year**2 - 22) + b
    p0 = expit(mu)
    p1 = expit(mu + log_psi)
    
    chain = np.zeros((nchain+1, 4)) # chaine de alpha, beta1, beta2 et sigma **2
    chain_b = np.zeros((nchain+1, K)) # chaine de b
    chain_mu = np.zeros((nchain+1, K)) # chaine de mu
    
    acc_rate = np.zeros(3)
    acc_rate_b = np.zeros(K)
    acc_rate_mu = np.zeros(K)
    
    # Initialisation de la chaîne
    chain_b[0,:] = initial_b
    chain_mu[0,:] = initial_mu
    chain[0,0] = initial_alpha
    chain[0,1] = initial_beta1
    chain[0,2] = initial_beta2
    chain[0,3] = initial_sigma
    

    for i in range(nchain):
        
        # Mise à jour de alpha
        proposition = normal(loc = alpha, scale = proposition_sd[0])
        proposition_log_psi = proposition + beta1 * year + beta2 * (year**2 - 22) + b
        proposition_p1 = expit(mu + proposition_log_psi)
        
        top = - (proposition**2 / (2 * 1e6)) + sum(r1 * np.log(proposition_p1)) + sum((n1-r1) * np.log(1 - proposition_p1))
        bottom = - (alpha**2 / (2 * 1e6)) + sum(r1 * np.log(p1)) + sum((n1-r1) * np.log(1 - p1))
        
        acc_prob = np.exp(top - bottom)
        
        if np.random.uniform() < acc_prob:
            alpha = proposition
            log_psi = proposition_log_psi
            p1 = proposition_p1
            acc_rate[0] += 1
        
        # Mise à jour de beta1
        proposition = normal(loc = beta1, scale = proposition_sd[1])
        proposition_log_psi = alpha + proposition * year + beta2 * (year**2 - 22) + b
        proposition_p1 = expit(mu + proposition_log_psi)
        
        top = - (proposition**2 / (2 * 1e6)) + sum(r1 * np.log(proposition_p1)) + sum((n1-r1) * np.log(1 - proposition_p1))
        bottom = - (beta1**2 / (2 * 1e6)) + sum(r1 * np.log(p1)) + sum((n1-r1) * np.log(1 - p1))
        
        acc_prob = np.exp(top - bottom)
        
        if np.random.uniform() < acc_prob:
            beta1 = proposition
            log_psi = proposition_log_psi
            p1 = proposition_p1
            acc_rate[1] += 1
            
        # Mise à jour de beta2
        proposition = normal(loc = beta2, scale = proposition_sd[2])
        proposition_log_psi = alpha + beta1 * year + proposition * (year**2 - 22) + b
        proposition_p1 = expit(mu + proposition_log_psi)
        
        top = - (proposition**2 / (2 * 1e6)) + sum(r1 * np.log(proposition_p1)) + sum((n1-r1) * np.log(1 - proposition_p1))
        bottom = - (beta2**2 / (2 * 1e6)) + sum(r1 * np.log(p1)) + sum((n1-r1) * np.log(1 - p1))
        
        acc_prob = np.exp(top - bottom)
        
        if np.random.uniform() < acc_prob:
            beta2 = proposition
            log_psi = proposition_log_psi
            p1 = proposition_p1
            acc_rate[2] += 1
            
        # Mise à jour de sigma
        sigma = 1/gamma(shape = 1e-3 + K/2, scale = 1e-3 + 0.5 * sum(b**2))
            
        for j in range(K):
            
            # Mise à jour de mu
            proposition = normal(loc = mu[j], scale = proposition_sd[3])
            proposition_p0_j = expit(proposition)
            proposition_p1_j = expit(proposition + log_psi[j])
            
            top = - (proposition**2 / (2 * 1e6)) + r0[j] * np.log(proposition_p0_j) + (n0[j] - r0[j]) * np.log(1 - proposition_p0_j) + r1[j] * np.log(proposition_p1_j) + (n1[j] - r1[j]) * np.log(1 - proposition_p1_j)
            bottom = - (mu[j]**2 / (2 * 1e6)) + r0[j] * np.log(p0[j]) + (n0[j] - r0[j]) * np.log(1 - p0[j]) + r1[j] * np.log(p1[j]) + (n1[j] - r1[j]) * np.log(1 - p1[j])
            
            acc_prob = np.exp(top - bottom)
            
            if np.random.uniform() < acc_prob:
                mu[j] = proposition
                p0[j] = proposition_p0_j
                p1[j] = proposition_p1_j
                acc_rate_mu[j] += 1
                
            # Mise à jour de b
            proposition = normal(loc = b[j], scale = proposition_sd[4])
            proposition_log_psi_j = alpha + beta1 * year[j] + beta2 * (year[j]**2 - 22) + proposition
            proposition_p1_j = expit(mu[j] + proposition_log_psi_j)
            
            top = - (proposition**2 / (2 * sigma)) + r1[j] * np.log(proposition_p1_j) + (n1[j] - r1[j]) * np.log(1 - proposition_p1_j)
            bottom = - (b[j]**2 / (2 * sigma)) + r1[j] * np.log(p1[j]) + (n1[j] - r1[j]) * np.log(1 - p1[j])
            
            acc_prob = np.exp(top - bottom)
            
            if np.random.uniform() < acc_prob:
                b[j] = proposition
                log_psi[j] = proposition_log_psi_j
                p1[j] = proposition_p1_j
                acc_rate_b[j] += 1
        
        # Mise à jour des chaines
        chain[i+1,:] = np.array([alpha, beta1, beta2, sigma])
        chain_b[i+1,:] = b 
        chain_mu[i+1,:] = mu
        
    return(chain,chain_b, chain_mu, acc_rate, acc_rate_b, acc_rate_mu)  
```

```python
chain, chain_b, chain_mu, acc_rate, acc_rate_b, acc_rate_mu = oxford(10000, initial_alpha, initial_beta1, initial_beta2, initial_sigma, initial_mu, initial_b, np.array([0.06, 0.03, 0.006, 0.1, 0.5]), K, r1, n1, r0, n0, year)
```
